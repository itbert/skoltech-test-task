### Алгоритм
Вы выбираете из группы из N друзей, с кем пойти на пикник. Однако друзья, связанные ребром на графе, находятся в очень недружественных отношениях друг с другом, поэтому, если они оба отправятся на пикник, он будет испорчен. Чтобы на пикнике было как можно больше друзей, кого следует пригласить?. Протестируйте алгоритм на случайных графах (5 - 10 штук)  различного размера, отрисуйте случайные графы и выведите финальный список на экран или в фаил.

```python
from utils import Graph
import random
```

```python
def find_max_independent_set(graph):
    remaining = set(graph.nodes.keys())
    independent_set = set()
    
    while remaining:
        node = min(remaining, key=lambda x: len(graph.adjacency[x] & remaining))
        independent_set.add(node)
        remaining -= graph.adjacency[node] | {node}
    
    return independent_set
```

```python

print(f'{10} примеров сгенерировано')

for i in range(10):
    print(f'№{i+1}')
    
    n = random.randint(8, 15)
    max_edges = n * (n - 1) // 2
    min_val = min(max_edges, 2 * n)
    e = random.randint(n, min_val)
    
    graph = Graph().generate_random_graph(n, e)
    
    ind_set = find_max_independent_set(graph)
    # friends_as_str: str = *list(sorted(ind_set))
    
    print(f'Итого: \
            \n\tУзлов (друзей): {n} \
            \n\tРебер (не пойдет со связанным): {e} \
            \n\tРазмер независ. сета (сколько друзей придет): {len(ind_set)} \
            \n\tВыбранные узлы (пригласить друзей из задачи): {sorted(ind_set)} \
        ')

    graph.draw(highlight_set=ind_set)
```
